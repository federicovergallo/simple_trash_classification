```python
import tensorflow as tf
```

```python
import numpy as np
import cv2
%matplotlib inline 
import matplotlib.pyplot as plt
import os
```

```python
IMG_HEIGHT, IMG_WIDTH = 224, 224
model_path = "checkpoints/mobilenet_aug_lite/Epoch_1000_model.hp5"
cans_test_path = "dataset/metal/" 
plastic_test_path = "dataset/plastic_web/"
set2class = [cans_test_path, "oranges", plastic_test_path]
```

```python
chosen_set = plastic_test_path
chosen_class = set2class.index(chosen_set)
```

```python
model = tf.keras.models.load_model(model_path)
```

```python
images_names = os.listdir(chosen_set)
images_names = [chosen_set+img for img in images_names]
```

```python
wrongs = []
for image_name in images_names:
    # Reading
    img = cv2.imread(image_name)[..., ::-1]
    img = cv2.resize(img, (IMG_HEIGHT, IMG_WIDTH), interpolation=cv2.INTER_AREA)
    img = np.expand_dims(img, axis=0)
    # Predicting 
    pred = np.argmax(model.predict(img))
    if pred != chosen_class:
        wrongs.append(image_name)
```

```python
accuracy = 1 - len(wrongs)/len(images_names)
accuracy
```
